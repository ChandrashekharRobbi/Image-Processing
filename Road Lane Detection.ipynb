# Importing the Libraries

```python
import matplotlib.pyplot as plt 
import matplotlib.image as img
import cv2
import numpy as np
```

# Viewing the Image

```python
def show_img(image):
    """
    Simple function to show the image by using matplotlib library
    """
    
    plt.imshow(image)
    plt.axis("off")
    plt.show();
```

```python
image = img.imread('test_image.jpg') 
show_img(image)
```

## Make the copy of the original one

```python
roadLaneImage = np.copy(image)
```

## Converting the image into the GrayScale Image

```python
grayScaleImage = cv2.cvtColor(roadLaneImage, cv2.COLOR_RGB2GRAY)
```

```python
cv2.imshow("Road Lane", grayScaleImage)
cv2.waitKey(10)
```

```python
cv2.destroyAllWindows()
```

## Reduction of Noise 
  Smoothing on top of grayscale images

```python
blur = cv2.GaussianBlur(grayScaleImage, (5,5), 0)
```

```python
show_img(blur)
```

```python
cv2.imshow("Road Lane", grayScaleImage)
cv2.waitKey()
```

```python
cv2.destroyAllWindows()
```
